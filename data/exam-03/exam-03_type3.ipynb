## 문제 3-3-1 (10점) — 단순선형회귀 (bmi → target)

`bmi` 를 독립변수, `target` 을 종속변수로 하는 **단순선형회귀모형** 을 적합하고,
회귀계수의 유의성을 검정하시오. (양측, α=0.05)

### 모형
$$ \text{target} = \beta_0 + \beta_1 \cdot \text{bmi} + \varepsilon $$

### 가설
- H₀: β₁ = 0 (bmi 는 target 에 영향 없음)
- H₁: β₁ ≠ 0 (bmi 는 target 에 영향 있음)

### 수행
- `statsmodels.api.OLS` 사용 — `sm.add_constant()` 로 상수항 포함

1. **기울기 β₁** (소수점 둘째 자리)
2. **t-통계량** (소수점 둘째 자리)

## 문제 3-3-2 (10점) — 다중회귀 계수 유의성 (bmi, bp, s5 → target)

`bmi`, `bp`, `s5` 세 변수를 독립변수로 하는 **다중회귀모형** 을 적합하고,
각 변수의 회귀계수가 target 에 미치는 영향이 통계적으로 유의한지 검정하시오. (양측, α=0.05)

### 모형
$$ \text{target} = \beta_0 + \beta_1 \cdot \text{bmi} + \beta_2 \cdot \text{bp} + \beta_3 \cdot \text{s5} + \varepsilon $$

### 가설
- 각 변수에 대해 H₀: β_j = 0 vs H₁: β_j ≠ 0 (j = 1, 2, 3)

### sklearn vs statsmodels 핵심 구분
| 라이브러리 | 주 목적 | 시험 영역 |
|---|---|---|
| `sklearn.linear_model.LinearRegression` | 예측 (RMSE, R²) | 작업형 2유형 |
| `statsmodels.api.OLS` | 추론 (t, p, F, summary) | **작업형 3유형** |

### statsmodels OLS 핵심 API
- `sm.add_constant(X)`: 절편 항 추가 (1열) — **반드시 필요**
- `model.params`: 회귀계수
- `model.tvalues`, `model.pvalues`: t-통계량 / p-value
- `model.fvalue`, `model.f_pvalue`: 전체 F-통계량 / p-value
- `model.rsquared`, `model.rsquared_adj`: R² / adjusted R²
- `model.summary()`: 전체 요약표

## 데이터 로드 (train.csv / test.csv)

````python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# 9회차 2유형 데이터 — 사이트 호스팅 train.csv / test.csv
BASE = 'https://raw.githubusercontent.com/ashram68/bigdata-analytics-exam/main/data/exam-09/'
train = pd.read_csv(BASE + 'train.csv')   # (142, 14) ID + 12변수 + alcohol
test  = pd.read_csv(BASE + 'test.csv')    # (36, 13)  ID + 12변수 (alcohol 미포함)
print('train:', train.shape)
print('test :', test.shape)
train.head()
````

**회귀 모델** 을 만드시오. 학습된 모델로 평가용 데이터(`test.csv`) 에 대한 **예측값** 을 산출하고,

4. 전체 학습 데이터로 재학습 후 `test` 예측값 산출 → `predict(X_test)`

## 풀이 종료 안내

### 회귀 vs 분류 result.csv 차이
- **회귀**: `predict()` 로 예측값(실수) 저장
- **분류**: `predict_proba()[:, 1]` 로 양성 확률(0~1) 저장 (AUC 평가 시)

### 회귀 성능 지표 해석
- **RMSE**: 단위가 타겟과 같음. alcohol 평균(~13.0) 대비 비율로 해석
- **R²**: 0~1 (음수도 가능). 0.7+ 우수, 0.5+ 양호 (빅분기 통과 기준)
- **MAE**: 이상치에 RMSE 보다 덜 민감, 절댓값 오차 평균

### 본 회차 3유형 핵심 — Pearson 상관 vs ANOVA

| 검정 | 통계량 | 효과 크기 | 적용 상황 |
|---|---|---|---|
| Pearson 상관 | r | r 자체 | 두 연속형 변수의 선형 관계 |
| 독립표본 t-검정 | t | Cohen's d | 2개 그룹 평균 차이 |
| One-way ANOVA | F | η² | 3개 이상 그룹 평균 차이 |
| 카이제곱 독립성 | χ² | Cramer's V | 두 범주형 변수 관련성 |

→ p-value 만 보지 말고 **효과 크기로 실질적 강도** 함께 평가하는 것이 실무 표준.

### ANOVA 응용
- η² 큰 변수 = ML 모델의 중요 변수 가능성 높음
- 사후 검정(post-hoc Tukey) 으로 어느 그룹 쌍이 유의한지 확인 가능
- 정규성 위반 시 Kruskal-Wallis (`stats.kruskal`) 사용

### 본 회차 3유형 핵심 — 가설 검정 + 효과 크기

| 검정 | 통계량 | 효과 크기 |
|---|---|---|
| 카이제곱 독립성 | χ² | Cramer's V |
| Pearson 상관 | r | r 자체 (절댓값) |
| 독립표본 t-검정 (참고) | t | Cohen's d |
| ANOVA (참고) | F | η² (eta-squared) |

→ p-value 만 보지 말고 효과 크기로 **실질적 강도** 함께 평가하는 것이 실무 표준.

## 데이터 로드 (train.csv / test.csv)

````python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# 8회차 2유형 데이터 — 사이트 호스팅 train.csv / test.csv
BASE = 'https://raw.githubusercontent.com/ashram68/bigdata-analytics-exam/main/data/exam-08/'
train = pd.read_csv(BASE + 'train.csv')   # (353, 12) ID + 10변수 + high_target
test  = pd.read_csv(BASE + 'test.csv')    # (89, 11)  ID + 10변수 (타겟 미포함)
print('train:', train.shape)
print('test :', test.shape)
train.head()
````

## 문제 8-2 (40점) — Diabetes 분류 모델 (high_target 예측)

학습용 데이터(`train.csv`) 의 10개 변수를 사용하여 `high_target`(0/1) 을 예측하는
**이진 분류 모델** 을 만드시오. 학습된 모델로 평가용 데이터(`test.csv`) 에 대한 **예측 확률** 을 산출하고,
그 결과를 `result.csv` 파일로 저장하시오.

### 수행 순서 (예시)
1. 결측치 처리: `age`, `bmi`, `s3` → `train` 의 중앙값 으로 train·test 동일하게 대체 (data leakage 방지)
2. 입력/타겟 분리: `X = train.drop(['ID', 'high_target'], axis=1)`, `y = train['high_target']`
3. 모델: `RandomForestClassifier(n_estimators=100, random_state=42)` 등 자유
4. 검증 분할 (선택): `train_test_split(test_size=0.2, random_state=42, stratify=y)` 로 검증 AUC 확인
5. 전체 학습 데이터로 재학습 후 `test` 예측 확률 산출 → `predict_proba(X_test)[:, 1]`
6. `result.csv` 저장: 헤더 포함, 컬럼 `ID, pred` (89행)

### 제출 형식
```
ID,pred
251,0.92
...
```

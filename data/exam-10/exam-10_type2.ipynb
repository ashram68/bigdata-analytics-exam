### RF vs GBM 차이
- **RF**: 병렬 학습, robust, 빠름, 기본 성능 우수
- **GBM**: 순차 학습 (이전 오류 보정), 더 정밀하나 overfitting 위험
- 작은 검증셋(n=91)에서는 두 모델 차이가 우연일 수 있음 → CV 로 확인 권장

### BC 데이터의 중요 변수
- `worst_perimeter`, `worst_concave_points`, `worst_radius` — 종양의 "최악" 부위 측정값
- `mean_concave_points`, `mean_radius` — 종양의 "평균" 측정값
- → 10-1-1 상관계수 절댓값 합 결과와 일관됨

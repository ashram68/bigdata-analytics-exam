### 통계 가정 점검 → 검정 선택 의사결정 트리

```
두 그룹 평균 비교가 필요한가?
│
├─ 정규성? (Shapiro)
│   ├─ 만족 → 등분산성? (Levene)
│   │       ├─ 만족 → 표준 t-test (ttest_ind, equal_var=True)
│   │       └─ 위반 → Welch's t-test (ttest_ind, equal_var=False)
│   └─ 위반 → 비모수: Mann-Whitney U (mannwhitneyu)
│
└─ 본 모의고사 결과:
    target=0 정규성 위반 + 등분산성 위반
    → Welch's t-test 또는 Mann-Whitney U 권장
```

### 정규성 검정 종류
- `stats.shapiro(data)` — Shapiro-Wilk (n<5000 권장)
- `stats.normaltest(data)` — D'Agostino-Pearson
- `stats.kstest(data, 'norm')` — Kolmogorov-Smirnov
- `stats.anderson(data, 'norm')` — Anderson-Darling

### 등분산성 검정 종류
- `stats.levene(*groups)` — Levene (정규성 robust, 권장)
- `stats.bartlett(*groups)` — Bartlett (정규성 가정 필요)
- `stats.fligner(*groups)` — Fligner-Killeen (정규성에 가장 robust)

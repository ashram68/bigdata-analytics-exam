## 문제 4-3-3 (10점) — Nested 모델 비교: bmi 추가 시 Adjusted R² 변화

[4-3-1] 에서 적합한 **Reduced 모델**(age + bp + s3 + s4) 에 `bmi` 변수를 추가한
**Full 모델**(age + bp + s3 + s4 + bmi) 을 적합하시오.
두 모델의 **Adjusted R²** 를 비교하여 bmi 추가가 모델 설명력을 의미있게 향상시키는지 판단하시오.

### 비교 모델
- **Reduced**: target = β₀ + β₁·age + β₂·bp + β₃·s3 + β₄·s4 + ε  (변수 4개)
- **Full**:    target = β₀ + β₁·age + β₂·bp + β₃·s3 + β₄·s4 + β₅·bmi + ε  (변수 5개)

### 출력 형식
1. **Reduced 모델 Adjusted R²** (소수점 넷째 자리)
2. **Full 모델 Adjusted R²** (소수점 넷째 자리)
3. **ΔAdjusted R²** (소수점 넷째 자리, 부호 포함)

- ΔAdj R² > 0.05 → bmi 추가 정당화
- ΔAdj R² ≈ 0 → 변수 추가 효과 미미
- ΔAdj R² < 0 → 변수 추가가 오히려 모델 악화

### 다중회귀 심화 3원칙 (4회차 핵심)
1. **단변량 상관 ≠ 다변량 부분효과** — 여러 변수 동시 고려해야 진짜 영향력 드러남
2. **회귀계수 크기 비교는 표준화된 변수에서만 유효** — 일반 데이터는 t-통계량 절댓값 사용
3. **변수 추가 정당화는 Adjusted R² 로** — 단순 R² 는 변수가 늘면 항상 증가

### Nested 모델 비교 방법
- Adjusted R² (본 문제) — 가장 직관적
- Partial F-test (`statsmodels.stats.anova.anova_lm`)
- AIC / BIC (정보 기준)

### 통계 검정 함수 빠른 참고
- 단일표본 평균 검정: `stats.ttest_1samp(data, popmean)`
- 독립표본 평균 비교: `stats.ttest_ind(g1, g2, equal_var=...)`
- 대응표본 평균 비교: `stats.ttest_rel(before, after)`
- 분산 동일성 (Levene): `stats.levene(g1, g2)`
- 정규성 (Shapiro): `stats.shapiro(data)`
- 일원분산분석 (3그룹↑): `stats.f_oneway(g1, g2, g3)`
- 적합성·독립성 (카이제곱): `stats.chi2_contingency(table)`
- 상관계수 (Pearson): `stats.pearsonr(x, y)`
- 로지스틱 회귀 + 오즈비: `LogisticRegression`, `OR = np.exp(coef)`

## 문제 2 (40점) — Titanic 생존자 예측 (이진 분류)

Titanic 데이터를 사용하여 승객의 **생존 여부(survived)** 를 예측하는 분류 모델을 만드시오.
다음 과정을 수행하고, 평가용 데이터에 대한 예측 결과를 `result.csv` 파일로 저장하시오.

### 수행 순서
1. 학습용/평가용 titanic 데이터 로드 (모의고사 플랫폼에서는 8:2 분할로 시뮬레이션)
2. 결측치 처리: `age` 는 **중앙값**, `embarked` 는 **최빈값**, `fare` 는 **중앙값** 으로 대체
3. 사용할 변수: `['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']`
4. 범주형 변수(`sex`, `embarked`) 는 `pd.get_dummies(drop_first=True)` 로 원-핫 인코딩
5. 학습용 데이터를 8:2 로 분할하여 검증셋 확보 (`random_state=42`, `stratify=y`)
6. `RandomForestClassifier(n_estimators=100, random_state=42)` 로 학습
7. 평가용 데이터의 **예측 확률** (`predict_proba`의 양성 클래스 [:, 1]) 로 ROC-AUC 계산
8. 또는 클래스 예측이 필요한 경우 `predict()` 로 0/1 예측
9. 예측 확률을 `result.csv` 파일로 저장

### 제출 형식
- 평가용 데이터의 **예측 확률** 을 `result.csv` 로 저장 (헤더 포함, `index=False`)
- 컬럼명: `pred` (또는 시험에서 지정한 이름)

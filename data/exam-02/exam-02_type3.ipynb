### 통계 검정 함수 빠른 참고
- 단일표본 평균 검정: `stats.ttest_1samp(data, popmean)`
- 독립표본 평균 비교: `stats.ttest_ind(g1, g2, equal_var=...)` ← 본 회차 사용
- 대응표본 평균 비교: `stats.ttest_rel(before, after)`
- 분산 동일성 (Levene): `stats.levene(g1, g2)` ← 본 회차 사용
- 정규성 (Shapiro): `stats.shapiro(data)`
- 일원분산분석: `stats.f_oneway(g1, g2, g3)`
- 적합성·독립성 (카이제곱): `stats.chi2_contingency(table)`
- **상관계수 (Pearson)**: `stats.pearsonr(x, y)` ← 본 회차 사용

### 효과 크기 해석 (Cohen's d / Pearson r)
- Cohen's d: 0.2(작음) / 0.5(중간) / 0.8(큼) / 1.2(매우 큼)
- Pearson r: |r|<0.3(약함) / 0.3~0.7(중간) / ≥0.7(강함)

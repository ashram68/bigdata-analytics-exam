## 데이터 로드

본 회차의 모든 문제는 `sklearn` 내장 **Iris** 데이터셋(150행 × 5컬럼)을 사용합니다.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import warnings
warnings.filterwarnings('ignore')

# 2회차 공통 데이터 — sklearn 내장 Iris (5컬럼 표준 구조)
iris = load_iris(as_frame=True).frame
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target']
iris['species'] = iris['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
iris = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']]

print(iris.shape)              # (150, 5)
print(iris['species'].value_counts())
iris.head()
```

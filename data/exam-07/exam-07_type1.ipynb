### 결측 비율별 처리 가이드
- **< 5%**: 단순 대체 (median, mean, mode) 또는 행 제거
- **5 ~ 30%**: 그룹별 대체, KNN 대체, 다중대체(MICE)
- **30 ~ 50%**: 신중한 대체 + 결측 표시 변수 추가 권장
- **> 50%**: 변수 제거 고려 또는 결측 자체를 정보로 활용 (본 문제 ca 가 해당)

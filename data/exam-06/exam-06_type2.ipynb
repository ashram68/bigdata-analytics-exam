## 풀이 종료 안내

### 모델 선택 기준
- **해석 가능성 중요** → Logistic Regression (계수 = 영향력)
- **비선형성/상호작용 강함** → RandomForest, XGBoost
- **빠른 baseline 필요** → Logistic Regression
- 시험에서는 두 가지 모두 시도 후 더 좋은 쪽 제출 가능

### StandardScaler 필수성
Logistic Regression 은 스케일에 민감 (특히 정규화 적용 시).
변수 단위가 다르면 큰 단위 변수가 회귀계수를 지배 → **반드시 스케일링**.
트리 모델(RF) 은 스케일 무관.

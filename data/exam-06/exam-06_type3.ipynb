(Age 결측치는 제거, 양측, α=0.05)

### 가설
- H₀: μ(생존자) = μ(사망자)
- H₁: μ(생존자) ≠ μ(사망자)

### 수행
Levene 등분산 검정 → `ttest_ind(equal_var=...)`

- Levene p < 0.05 → 이분산 (`equal_var=False`, Welch's t-test)

### 출력 형식
1. **t-통계량** (소수점 둘째 자리)

### 본 회차 3유형 핵심 — t-검정 vs 카이제곱

| 검정 | 통계량 | 효과 크기 | 적용 상황 |
|---|---|---|---|
| 독립표본 t-검정 | t | Cohen's d | 두 그룹 평균 차이 (연속 vs 범주) |
| 카이제곱 독립성 | χ² | Cramer's V | 두 범주형 변수 연관성 |

### 카테고리별 가설 검정 가이드
- **두 그룹 평균 비교 (연속 vs 범주)**: 독립표본 t-검정 (정규성 만족), Mann-Whitney U (비모수)
- **두 범주형 변수 연관성**: 카이제곱 독립성 검정, Fisher's exact test (소표본)
- **두 연속형 변수 관계**: Pearson 상관 (선형), Spearman (순위)
- **3그룹 이상 평균 비교**: ANOVA + Tukey HSD
- **변수 영향력 (회귀)**: OLS (연속 종속변수), Logit (이진 종속변수)
